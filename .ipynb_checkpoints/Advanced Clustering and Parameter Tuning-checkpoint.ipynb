````python
# Import Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN, KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score, adjusted_rand_score, davies_bouldin_score
from sklearn.datasets import make_blobs, load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
````

### Load dataset

````python
data = load_iris()
X = data.data
y_true = data.target  # True labels for evaluation
````

````python
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
````

## Visualize the dataset in 2D using PCA

````python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(6, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], s=30, c='blue', alpha=0.7)
plt.title("Iris Dataset (PCA Reduced to 2D)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
````

## Parameter Tuning for Clustering Algorithms

### Tuning DBSCAN

````python
# Experiment with different `eps` and `min_samples` values
eps_values = [0.5, 0.7, 1.0]
min_samples_values = [3, 5, 10]

# Store results for evaluation
dbscan_results = {}

for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X_scaled)
        silhouette = silhouette_score(X_scaled, labels) if len(set(labels)) > 1 else -1
        dbscan_results[(eps, min_samples)] = silhouette

# Display results
print("DBSCAN Parameter Tuning Results (Silhouette Score):")
for params, score in dbscan_results.items():
    print(f"eps={params[0]}, min_samples={params[1]}: Silhouette Score={score:.4f}")
````

### Tuning K-Means

````python
# Experiment with different values of `k`
k_values = range(2, 10)
kmeans_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    silhouette = silhouette_score(X_scaled, labels)
    kmeans_scores.append(silhouette)

# Plot K-Means Silhouette Scores
plt.figure(figsize=(6, 6))
plt.plot(k_values, kmeans_scores, marker='o', linestyle='-', color='b')
plt.title("K-Means Silhouette Scores")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.show()
````

````python
# Experiment with different values of `k`
k_values = range(2, 10)
kmeans_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    silhouette = silhouette_score(X_scaled, labels)
    kmeans_scores.append(silhouette)

# Find the best `k` based on the highest Silhouette Score
best_k_index = kmeans_scores.index(max(kmeans_scores))
best_k = k_values[best_k_index]
best_score = kmeans_scores[best_k_index]

# Plot K-Means Silhouette Scores
plt.figure(figsize=(6, 6))
plt.plot(k_values, kmeans_scores, marker='o', linestyle='-', color='b', label='Silhouette Scores')
plt.scatter([best_k], [best_score], color='red', s=100, label=f'Best k={best_k} (Score={best_score:.2f})')
plt.title("K-Means Silhouette Scores")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.legend()
plt.grid(True)
plt.show()
````

### Why Maximize Silhouette Score?

- The **Silhouette Score** measures how well each point is clustered:
  - A score close to **1** indicates that a sample is well-matched to its own cluster and poorly matched to neighboring clusters.
  - A score close to **0** indicates that the sample lies on or very near the decision boundary between two clusters.
  - A score close to **-1** indicates that the sample has been assigned to the wrong cluster.

Thus, higher Silhouette Scores indicate better-defined and more compact clusters, which is the goal of clustering algorithms.

### Summary

To select the optimal number of clusters (\(k\)) or evaluate the performance of a clustering algorithm:
- **Maximize the Silhouette Score**.


### Evaluate Clustering Performance

````python
# Use the best parameters for evaluation

# DBSCAN (best eps=0.7, min_samples=5 based on previous results)
dbscan = DBSCAN(eps=0.7, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)

# K-Means (best k=3 based on Silhouette Score)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

# Hierarchical Agglomerative Clustering (HAC)
hac = AgglomerativeClustering(n_clusters=3)
hac_labels = hac.fit_predict(X_scaled)
````

````python
# Evaluation Metrics
print("\nClustering Performance Evaluation:")
print(f"DBSCAN - Silhouette Score: {silhouette_score(X_scaled, dbscan_labels):.4f}")
print(f"K-Means - Silhouette Score: {silhouette_score(X_scaled, kmeans_labels):.4f}")
print(f"HAC - Silhouette Score: {silhouette_score(X_scaled, hac_labels):.4f}")

print(f"\nAdjusted Rand Index (vs. true labels):")
print(f"DBSCAN - ARI: {adjusted_rand_score(y_true, dbscan_labels):.4f}")
print(f"K-Means - ARI: {adjusted_rand_score(y_true, kmeans_labels):.4f}")
print(f"HAC - ARI: {adjusted_rand_score(y_true, hac_labels):.4f}")

print(f"\nDavies-Bouldin Index (lower is better):")
print(f"DBSCAN - DBI: {davies_bouldin_score(X_scaled, dbscan_labels):.4f}")
print(f"K-Means - DBI: {davies_bouldin_score(X_scaled, kmeans_labels):.4f}")
print(f"HAC - DBI: {davies_bouldin_score(X_scaled, hac_labels):.4f}")
````

### Visualizing Clusters in 2D

````python
# Plot clustering results for K-Means
plt.figure(figsize=(6, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans_labels, cmap='viridis', s=30, alpha=0.7)
plt.title("K-Means Clustering (PCA Reduced to 2D)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label="Cluster")
plt.show()

# Plot clustering results for DBSCAN
plt.figure(figsize=(6, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=dbscan_labels, cmap='viridis', s=30, alpha=0.7)
plt.title("DBSCAN Clustering (PCA Reduced to 2D)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label="Cluster")
plt.show()

# Plot clustering results for HAC
plt.figure(figsize=(6, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=hac_labels, cmap='viridis', s=30, alpha=0.7)
plt.title("HAC Clustering (PCA Reduced to 2D)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label="Cluster")
plt.show()
````

### Explanation of Clustering Metrics

#### 1. **Silhouette Score**
The **Silhouette Score** measures how similar a data point is to its own cluster (cohesion) compared to other clusters (separation). It ranges from `-1` (incorrect clustering) to `+1` (ideal clustering). A value near `0` indicates overlapping clusters.

**Equation**:

$\
s(i) = \frac{b(i) - a(i)}{\max(a(i), b(i))}
\$

Where:
- \(a(i)\): Mean distance between a sample \(i\) and all other points in the same cluster.
- \(b(i)\): Mean distance between a sample \(i\) and all points in the nearest cluster.


---

#### 2. **Adjusted Rand Index (ARI)**
The **ARI** measures the similarity between two clusterings by comparing the number of data point pairs assigned to the same or different clusters in both clusterings. It is adjusted for chance and ranges from `-1` (completely dissimilar) to `1` (perfect match).

**Equation**:
$
\text{ARI} = \frac{\text{Index} - \text{Expected Index}}{\text{Max Index} - \text{Expected Index}}
$
- **Index**: Number of agreements between predicted and true labels.
- **Expected Index**: Expected number of agreements by chance.
- **Max Index**: Maximum possible number of agreements.

---

#### 3. **Davies-Bouldin Index (DBI)**
The **DBI** evaluates clustering quality by analyzing intra-cluster variance and inter-cluster separation. Lower values indicate better clustering.

**Equation**:
$
DBI = \frac{1}{k} \sum_{i=1}^{k} \max_{j \neq i} \left( \frac{\sigma_i + \sigma_j}{d_{ij}} \right)
$
- $(k)$: Number of clusters.
- $(\sigma_i)$: Average distance of points in cluster \(i\) to the centroid of cluster \(i\).
- $(d_{ij})$: Distance between centroids of clusters \(i\) and \(j\).

---

### Comparison Table

| Metric                 | Range           | Purpose                                           | Best Value | Equation Highlights                          |
|------------------------|-----------------|---------------------------------------------------|------------|---------------------------------------------|
| **Silhouette Score**   | `-1` to `1`    | Measures cohesion and separation of clusters.    | Close to 1 | Uses mean distances within and between clusters. |
| **Adjusted Rand Index (ARI)** | `-1` to `1` | Compares predicted vs. true labels.              | Close to 1 | Considers agreements between clusterings.    |
| **Davies-Bouldin Index (DBI)** | `0` to       | Evaluates intra-cluster variance and separation. | Close to 0 | Balances cluster compactness and separation. |


---

### Key Differences:
1. **Silhouette Score**:
   - Unsupervised: Does not require ground truth labels.
   - Good for understanding the overall clustering quality.

2. **ARI**:
   - Supervised: Requires true labels to compare against.
   - Measures clustering accuracy and agreement.

3. **DBI**:
   - Unsupervised: Focuses on cluster compactness and separation.
   - Lower values indicate better cluster structures.
