### Import necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, Normalizer
from sklearn.decomposition import PCA
```

### Step 1: Generate synthetic data

```python
np.random.seed(42)
data = np.random.rand(100, 3) * [10, 100, 1000]  # Features with different scales
columns = ['Feature_1', 'Feature_2', 'Feature_3']
df = pd.DataFrame(data, columns=columns)
```

### Step 2: Apply scaling and normalization methods

```python
transformations = {
    'Original Data': df,
    'Min-Max Scaling': pd.DataFrame(MinMaxScaler().fit_transform(df), columns=columns),
    'Standard Scaling': pd.DataFrame(StandardScaler().fit_transform(df), columns=columns),
    'Robust Scaling': pd.DataFrame(RobustScaler().fit_transform(df), columns=columns),
    'Normalization': pd.DataFrame(Normalizer().fit_transform(df), columns=columns)
}
```

### Step 3: Perform PCA for visualization (reduce to 2D)

```python
pca_results = {}
for name, transformed_data in transformations.items():
    pca = PCA(n_components=2)
    pca_results[name] = pca.fit_transform(transformed_data)
```

### Step 4: Visualize results

```python
fig, axes = plt.subplots(3, 2, figsize=(15, 15))
axes = axes.ravel()

for ax, (name, pca_data) in zip(axes, pca_results.items()):
    ax.scatter(pca_data[:, 0], pca_data[:, 1], alpha=0.7)
    ax.set_title(name)
    ax.set_xlabel('PCA Component 1')
    ax.set_ylabel('PCA Component 2')

# Adjust layout for clarity
plt.tight_layout()
plt.show()
```

## Load the Titanic dataset

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

### Select a subset of columns for simplicity

```python
df = df[['Survived', 'Pclass', 'Age', 'Fare', 'Embarked']]

# Handle missing values
df['Age'] = df['Age'].fillna(df['Age'].mean())  # Replace missing 'Age' with mean
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])  # Replace missing 'Embarked' with mode

# Encode categorical features ('Embarked')
df['Embarked'] = df['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
```

## Data Manipulation

```python
# Separate features from the target
features = df[['Pclass', 'Age', 'Fare', 'Embarked']]
```

### Transformation: Min-Max Scaling (Rescale features to [0, 1])


```python
scaler = MinMaxScaler()
features_transformed = pd.DataFrame(scaler.fit_transform(features), columns=features.columns)
```

### Normalization: Normalize each sample (row) to have unit norm


```python
normalizer = Normalizer()
features_normalized = pd.DataFrame(normalizer.fit_transform(features), columns=features.columns)
```

```python
# Summary of Results
print("Original Data (First 5 Rows):\n", features.head())
print("\nTransformed Data (Min-Max Scaled, First 5 Rows):\n", features_transformed.head())
print("\nNormalized Data (Row Norm = 1, First 5 Rows):\n", features_normalized.head())
```

### Visualization of Data Differences

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)
axes[0].boxplot(features.values, vert=True, patch_artist=True)
axes[0].set_title('Original Data')

axes[1].boxplot(features_transformed.values, vert=True, patch_artist=True)
axes[1].set_title('Transformed Data (Min-Max Scaling)')

axes[2].boxplot(features_normalized.values, vert=True, patch_artist=True)
axes[2].set_title('Normalized Data')

for ax in axes:
    ax.set_xticks(range(1, 5))
    ax.set_xticklabels(features.columns)

plt.tight_layout()
plt.show()
```

### Visualize Histograms for Original, Transformed, and Normalized Data


```python
fig, axes = plt.subplots(4, 3, figsize=(18, 16), sharey=True)

# Titles for the rows
titles = ['Original Data', 'Transformed Data (Min-Max Scaling)', 'Normalized Data']
datasets = [features, features_transformed, features_normalized]

# Iterate over each feature and create histograms
for i, column in enumerate(features.columns):
    for j, data in enumerate(datasets):
        axes[i, j].hist(data[column], bins=20, alpha=0.7, edgecolor='black')
        axes[i, j].set_title(f"{titles[j]} - {column}")
        axes[i, j].set_xlabel(column)
        axes[i, j].set_ylabel("Frequency")

# Adjust layout
plt.tight_layout()
plt.show()
```
